LINEAR REGRESSION MODEL

```python
from sklearn import datasets, linear_model
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
X =house[['bedroom_count' , 'Population' , 'Distance from CBD' , 'Hospital' , 'School' , 'Super Market']]

Y = house['sold_price']
```

```python
X
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 20, random_state= 0)
```

```python
X_train['bedroom_count']
```

```python
regressor = LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
 y_pred = regressor.predict(X_test)
```

```python
errors = abs(y_pred - Y_test)
mape = 100 * (errors / Y_test)     #mean absolute percentage error
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```

```python
print('MAE:',metrics.mean_absolute_error(Y_test,y_pred))
print('MSE:',metrics.mean_squared_error(Y_test,y_pred))
print('RMSE:',np.sqrt(metrics.mean_squared_error(Y_test,y_pred)))
print('R^2 =',metrics.explained_variance_score(Y_test,y_pred))
```

```python
plt.scatter(Y_test, y_pred)
```

RANDOM FOREST AND PREDICTIONS

```python
house['method'] = house['Method'].astype('category').cat.codes    #convert categorical to numerical for modelling
house['seller'] = house['agency_name'].astype('category').cat.codes
house['suburb'] = house['SUBURB'].astype('category').cat.codes
house['type'] = house['dwelling_type'].astype('category').cat.codes 
```

```python
house['method']
```

```python
X =house[['bedroom_count' , 'Population' , 'Distance from CBD' , 'Hospital' , 'School' , 'Super Market' , 'method' , 'seller' , 'suburb' , 'type']]    # X includes the independent variable and Y depicts dependent variable

Y = house['sold_price'] 
```

```python
X
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 20, random_state= 12)  #training and trsting the data
#standard test:train ratio is taken as 80:20
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)   # n depicts the number of iterations
rf.fit(X_train,Y_train);    #implementing random forest model
```

```python
predictions = rf.predict(X_test)
errors = abs(predictions - Y_test)
```

```python
mape = 100 * (errors / Y_test)     #mean absolute percentage error
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')   #to find the model accuracy
```

```python
from sklearn import metrics
print('r^2 =',metrics.explained_variance_score(Y_test,predictions))    #to depict R^2 value
```

PREDICTIONS

```python
plt.scatter(Y_test,predictions)  
```

```python
import seaborn as sn
confusion_matrix = pd.crosstab(Y_test, predictions, rownames=['Actual'], colnames=['Predicted'])
sn.heatmap(confusion_matrix, annot=True)
```

```python
feature_labels = np.array(['bedroom_count' , 'Population' , 'Distance from CBD' , 'Hospital' , 'School' , 'Super Market' , 'method' , 'seller' , 'suburb' , 'type'])
importance = rf.feature_importances_
feature_indexes_by_importance = importance.argsort()
for index in feature_indexes_by_importance:
    print('{}-{:.2f}%'.format(feature_labels[index], (importance[index] *100.0)))
    
sn.barplot(x=feature_labels, y=importance)
plt.xlabel('Features Importance')
plt.show()
```
